```python
import tensorflow as tf
import os
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import csv
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,horizontal_flip = True)
training_set = train_datagen.flow_from_directory(
    '../uTHCD_a(70-30-split)/70-30-split/train-test-classwise/Train',
    target_size = (64, 64),
    batch_size = 32,
    class_mode = 'categorical'
)
```

```python
model = tf.keras.models.load_model('h5 basicArc.h5')
```

```python
def insertData(data):
    with open("output.txt","a") as f:
        f.write(f'{data}\n')
```

```python
def getKey(value):
    dictionary = training_set.class_indices
    return list(dictionary.keys())[list(dictionary.values()).index(value)]
```

```python
def prediction(path):
    test_image = image.load_img(path, target_size = (64, 64))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model.predict(test_image)
    class_type = np.where(result[0] == 1.)[0]
    return class_type
```

```python
def singleInput(path):
    val = prediction(path)
    print(val)
```

```python
def multipleInputs(path):
    entries = os.listdir(path)
    n_true = 0
    n_false = 0
    print("Input  :  Output  :  Prediction")
    print()
    for i in entries:
        val = prediction(f'{path}/{i}')
        out = 0
        try:out = getKey(int(val[0]))
        except:pass
        acc = int(out) == int(i[-7:-4])
        if acc:n_true += 1
        else:n_false += 1
        print(f'class {i[-7:-4]}', "  :  ", f'class {out}',"  :  ",acc)
    print()
    print(f"Total Number of Inputs : {len(entries)}")
    print(f"True Predictions : {n_true}")
    print(f"False Predictions : {n_false}")
    print(f"accuracy : {(n_true/len(entries))*100}")
```

```python
path = "../uTHCD_d(90-10-split)/90-10-split/Test/"
multipleInputs(path)
```
